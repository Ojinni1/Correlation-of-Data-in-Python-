# Data Correlation in python 

In this Project, i import a dataset(movies.csv) using read_csv into the jupyter notebook, cleansed it by adjusting the Dates, dropping of Nan values and dropping duplicates. before working on the data correlation.

```python
#importing our library for the project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None # size of the plot to be use
```

```python
# importing the data(movies.csv dataset) to be use and also assigning a variable to it 

movies = pd.read_csv(r'/Users/gokeojinni/Downloads/movies.csv')

movies     # previewing the data  
```

```python
# checking for how many  NaN value in the data
movies.isnull().sum()
```

```python
'''to drop all NaN values in our datframe because we can't replace a budget with 0 nor the gross itll skew our 
result the best is to drop it '''

movies = movies.dropna(how= 'any', axis = 0)
```

```python
# checking our dataframe if we successfully drop the NaN values 
movies.isnull().sum()
```

```python
# checking for the data types we're having in the dataframe

movies.dtypes
```

```python
#working with budget and gross columns as float might be a little bit inconvenience needs to change it to an int
movies['budget'] = movies['budget'].astype(int)
movies['gross'] = movies['gross'].astype(int)
movies
```

```python
#the released date and the year doesn't match at some point so had to use the released date as the actual date 

movies['yearcorrect'] = movies['released'].str.extract(pat = '([0-9]{4})').astype(int)

movies = movies.drop('year', axis = 1)
```

```python
movies.head(3)
```

```python
movies = movies.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
#in case of any duplicates, it needs to to drop 

movies = movies.drop_duplicates()
```

```python
movies.head(5)
```

```python
# using scatter plot 

plt.scatter(x=movies['budget'], y=movies['gross'])
plt.title('Budget vs. Gross Earning')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earning')

plt.show
```

```python
movies.head(3)
```

```python
# using seaborn to check for the correlation 

sns.regplot(x='budget', y= 'gross', data= movies, scatter_kws={'color' : 'red'}, line_kws={'color': 'blue'})
```

```python
#looking at the correlation 
```

Correlation is a statistical techniques which helps inn analyziong the relationship between two or more variables

```python
movies.corr(method = 'pearson') other methods that can be used #kendall, spearman 
```

```python
correlation_matrix = movies.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
#checking for companies
movies.head()
```

converting the columns with strings 

```python
movies_numerized = movies.copy()

for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes # it'll give it a random numerization 

movies_numerized
```

Correleation for every field in the movies dataset 

```python
correlation_matrix = movies_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
movies_numerized.corr()
```

```python
correlation_mat = movies_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)> 0.5]

high_corr
```
